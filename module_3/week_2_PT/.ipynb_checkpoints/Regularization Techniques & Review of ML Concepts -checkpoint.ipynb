## Review of important ML Concepts

### Bias variance tradeoff

![](http://scott.fortmann-roe.com/docs/docs/BiasVariance/biasvariance.png)

![](https://miro.medium.com/max/544/1*Y-yJiR0FzMgchPA-Fm5c1Q.jpeg)

1. **High bias**<br>
    a. Systematic error in predictions (ie the average)<br>
    b. Bias is about the strength of assumptions the model makes<br>
    c. Underfit models tend to be high bias
    
2. **High variance**<br>
    a. The model is highly sensitive to changes in the data<br>
    b. Overfit models tend to be low bias

### Overfitting vs Underfitting

![](https://gblobscdn.gitbook.com/assets%2F-LvBP1svpACTB1R1x_U4%2F-LvNWUoWieQqaGmU_gl9%2F-LvNoby-llz4QzAK15nL%2Fimage.png?alt=media&token=41720ce9-bb66-4419-9bd8-640abf1fc415)
1. **Underfit Models fail to capture all of the information in the data**<br>
    1. Ex. the mean leaves a lot of information on the table
    
    
2. **Overfit models fit to the noise in the data and fail to generalize**


3. **How would we know if our model is over or underfit?**<br>
    1. Train test split<br>
    2. Look at the testing error<br>
    3. As model complexity increases so does the possibility for overfitting

## Regularization 

Linear models tend to have high bias but as you add more features, create more interactions and transformations the bias gets decreased, variance increases and models begin to overfit. You can then introduce L1(LASSO) or L2(RIDGE) regularization techniques to decrease the overfitting. But how do we do that? 

### When should you use L1 and L2 Regularization? 
* **L2**
 - when you have features with high multicollinearity
 - reduce model complexity
 - features with large coefficients
 
* **L1**
 - when you have a lot of small coefficients
 - when you have LOTS of features

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import mean_squared_error
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ads = pd.read_csv('Advertising.csv', index_col=0)
```

```python
ads.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(ads[['TV',
                                                         'Radio',
                                                         'Newspaper']],
                                                   ads['Sales'])
```

```python
sns.pairplot(ads);
```

```python
lr1 = LinearRegression()
lr1.fit(X_train, y_train)

ads_preds = lr1.predict(X_test)

np.sqrt(mean_squared_error(ads_preds, y_test))
```

```python
ss = StandardScaler()
pf = PolynomialFeatures()

X_train_processed = pf.fit_transform(ss.fit_transform(X_train))
X_test_processed = pf.transform(ss.transform(X_test))

#alpha == lambda since lamda is a key word in python
rr = Ridge(alpha=1)

rr.fit(X_train_processed, y_train)
ridge_preds = rr.predict(X_test_processed)

np.sqrt(mean_squared_error(ridge_preds, y_test))
```

```python
from sklearn.linear_model import RidgeCV
clf = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1]).fit(X_train, y_train)
clf.score(X_train, y_train)
```
